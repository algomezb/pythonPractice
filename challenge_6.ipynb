# Plots

```python
import pandas as pd
# Note: Update the relative path to the data
my_data = pd.read_csv("tmdb-movies.csv")
```

1. Create a plot of average runtime from year to year.

```python
my_data.groupby('release_year')['runtime'].mean().plot()
```

2. Create a plot of the average popularity of movies for each year.

```python
my_data.groupby('release_year')['popularity'].mean().plot()
```
